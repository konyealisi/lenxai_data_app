```python
# Read CSV file
csv_file = "C:/Users/konye/Downloads/dataUpload.csv"
df = pd.read_csv(csv_file)

# # Remove '?' from column names
# df.columns = df.columns.str.replace('?', '')

# Remove text within brackets (including the brackets) and question marks from column names
df.columns = df.columns.str.replace(r'\s*[\(\[].*?[\)\]]|\?', '', regex=True)

# Remove anything after a slash (including the slash) from column names
df.columns = df.columns.str.split('/').str[0]
```

```python
def clean_dataframe(df):
    # Trim column names, convert to lowercase, and replace spaces with underscores
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
    
    # Trim all string values in the DataFrame
    df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
    
    return df
clean_df = clean_dataframe(df)
df = clean_df.drop('s', axis=1)
df
```

```python
import pandas as pd
import numpy as np

# Read the data
data = pd.read_csv('facility_data.csv')

# Calculate proportions
data['total_tx'] = data.groupby(['Region', 'Funder', 'Implementing Partner', 'Implemention Type'])['TX_CURR'].transform('sum')
data['proportion'] = data['total_tx'] / data['TX_CURR'].sum()

# Target TX_CURR values for each group
target_total = 200000
data['target_tx'] = data['proportion'] * target_total

# Function to sample facilities based on the target TX_CURR
def sample_facilities(group, target_tx):
    target = group['target_tx'].iloc[0]
    sampled = pd.DataFrame(columns=group.columns)

    while target > 0 and not group.empty:
        idx = np.random.choice(group.index, replace=False)
        selected = group.loc[idx]
        group = group.drop(idx)

        if target - selected['TX_CURR'] >= 0:
            sampled = sampled.append(selected)
            target -= selected['TX_CURR']
    
    return sampled

# Function to check the sampling conditions
def check_conditions(sampled_data, target_total, tolerance, min_states, max_states):
    total_tx = sampled_data['TX_CURR'].sum()
    unique_states = len(sampled_data['State'].unique())

    lower_bound = target_total * (1 - tolerance)
    upper_bound = target_total * (1 + tolerance)

    return (lower_bound < total_tx <= upper_bound) and (min_states < unique_states <= max_states)

# Loop until the conditions are met
while True:
    sampled_data = data.groupby(['Region', 'Funder', 'Implementing Partner', 'Implemention Type']).apply(sample_facilities)

    if check_conditions(sampled_data, target_total, 0.05, 10, 15):
        print("Sampling conditions met.")
        sampled_data.to_csv('sampled_facilities.csv', index=False)
        break
    else:
        print("Sampling conditions not met. Trying again...")
```
