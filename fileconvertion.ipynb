```python
import pandas as pd
import re
```

```python
# Read CSV file
csv_file = "C:/Users/konye/Downloads/dataUpload.csv"
df = pd.read_csv(csv_file)

# # Remove '?' from column names
# df.columns = df.columns.str.replace('?', '')

# Remove text within brackets (including the brackets) and question marks from column names
df.columns = df.columns.str.replace(r'\s*[\(\[].*?[\)\]]|\?', '', regex=True)

# Remove anything after a slash (including the slash) from column names
df.columns = df.columns.str.split('/').str[0]
```

```python
def clean_dataframe(df):
    # Trim column names, convert to lowercase, and replace spaces with underscores
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
    
    # Trim all string values in the DataFrame
    df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
    
    return df
clean_df = clean_dataframe(df)
df = clean_df.drop('s', axis=1)
df
```

```python
# Convert DataFrame to XML
# df1 = df.copy()
# df1 = df1.drop('S/No.', axis=1)
xml_output_file = 'C:/Users/konye/Downloads/dataUpload.xml'
df.to_xml(xml_output_file, index=False, root_name='Root', row_name='Row')
```

```python
# Convert DataFrame to JSON
json_output_file = 'C:/Users/konye/Downloads/dataUpload.json'
df.to_json(json_output_file, orient='records', date_format='iso')
```
