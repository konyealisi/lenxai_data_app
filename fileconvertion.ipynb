```python
from datetime import timedelta
import pandas as pd

def curr(last_pickup_date, days_of_arv_refill, cutoff, grace_period):
    if last_pickup_date is pd.NaT or last_pickup_date is None:
        return None

    days_of_arv_refill_t = timedelta(days=days_of_arv_refill)
    grace_period_timedelta = timedelta(days=grace_period)

    if (last_pickup_date + days_of_arv_refill_t + grace_period_timedelta).date() >= pd.Timestamp(cutoff).date():
        return "yes"
    else:
        return "no"

cutoff = '2022-12-31'
grace_period = 28

data['days_of_arv_refill'] = data['days_of_arv_refill'].fillna(0).astype(int)
data['last_drug_pickup_date'] = pd.to_datetime(data['last_drug_pickup_date'], errors='coerce')

data['Cal_CURR'] = data.apply(lambda row: curr(row['last_drug_pickup_date'], row['days_of_arv_refill'], cutoff, grace_period), axis=1)
```
